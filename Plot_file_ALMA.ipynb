```
from matplotlib import pyplot as plt, cm 
from matplotlib import colors
from astropy.visualization import astropy_mpl_style
plt.style.use(astropy_mpl_style)
import numpy as np

from astropy.utils.data import get_pkg_data_filename
from astropy.io import fits

from astropy.visualization import imshow_norm, ZScaleInterval

from google.colab import drive #Mounting google drive. Insert confirmation code.
drive.mount('/content/drive')
image_file = get_pkg_data_filename('/content/drive/My Drive/M12_1092_continuum.fits')

fits.info(image_file)

image_data = fits.getdata(image_file, ext=0)
image_header = fits.getheader(image_file, ext=0)

display(image_header)

print(image_data.shape)

i=image_data [0,0,:,:]
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
im, norm = imshow_norm(i, ax, origin='lower',interval=ZScaleInterval())
fig.colorbar(im)
#print(i.shape)
#print(np.argmax(i, axis=0))
```
